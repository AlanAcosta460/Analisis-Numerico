### Ecuaciones algebraicas y trascendentales

```python
red = 4
tol = 0.0001
max_iter = 25

def biseccion() :
    print("*** Bisección ***")
    print("La función es: 0.5x^3 - 2x^2 + 3x - 90 = 0")

    print(f"Con tolerancia = {tol} y maximo de iteraciones = {max_iter}\n")
    tabular1(4)

    while True:
        xI = float(input("\n*** Ingresa el valor inicial xI *** \n$ "))
        xS = float(input("\n*** Ingresa el valor final xS *** \n$ "))

        fxI = f(xI)
        fxS = f(xS)

        if fxI * fxS >= 0 :
            print("\n\tEl intervalo no es valido\n")
        else :
            break

    print("\n\ni\txI\t\txP\t\txS\t\tf(xI)\t\tf(xP)\t\tf(xS)\t\t|xS - xI|\tvs tol")
    i = 1
    diff = 1
    while diff <= tol or i <= max_iter:
        xP = round((xI + xS) / 2, red)
        fxI = f(xI)
        fxS = f(xS)
        fxP = f(xP)

        dif = round(abs(xS - xI), red)
        print(f"{i}\t{xI}\t\t{xP}\t\t{xS}\t\t{fxI}    \t{fxP}    \t{fxS}\t\t{dif}\t\t{'Cumple' if dif <= tol else 'No cumple'}")

        if fxI * fxP < 0 :
            xS = xP
        else :
            xI = xP

        i += 1

    print(f"\n*** Raiz = {xP} ***")


def falsa_posicion() :
    print("*** Falsa Posición ***")
    print("La función es: 0.5x^3 - 2x^2 + 3x - 90 = 0")

    print(f"Con tolerancia = {tol} y maximo de iteraciones = {max_iter}\n")
    tabular1(4)

    while True :
        xI = float(input("\n*** Ingresa el valor inicial xI *** \n$ "))
        xS = float(input("\n*** Ingresa el valor final xS *** \n$ "))

        fxI = f(xI)
        fxS = f(xS)

        if fxI * fxS >= 0 :
            print("\n\tEl intervalo no es valido\n")
        else :
            break

    print("\n\ni\txI\t\txP\t\txS\t\tf(xI)\t\tf(xP)\t\tf(xS)\t\t|xS - xI|\tvs tol")
    i = 1
    diff = 1
    while diff <= tol or i <= max_iter:
        xP = round(xS - (fxS * (xI - xS)) / (fxI - fxS), red)
        fxI = f(xI)
        fxS = f(xS)
        fxP = f(xP)

        dif = round(abs(xS - xI), red)
        print(f"{i}\t{xI}\t\t{xP}\t\t{xS}\t\t{fxI}    \t{fxP}    \t{fxS}\t\t{dif}\t\t{'Cumple' if dif <= tol else 'No cumple'}")

        if fxI * fxP < 0 :
            xS = xP
        else :
            xI = xP

        if dif <= tol :
            break

        i += 1

    print(f"\n*** Raiz = {xP} ***")


def newton_raphson() :
    print("*** Newton-Raphson ***")
    print("La función es: 0.5x^3 - 2x^2 + 3x - 90 = 0")

    print(f"Con tolerancia = {tol}, maximo de iteraciones = {max_iter}\n")
    tabular1(4)

    while True:
        try:
            x0 = float(input("\n*** Ingresa el valor inicial *** \n$ "))
            
            criterio = G_1(x0)
            print("\n\n*** Prueba de Convergencia ***")
            print(f"x0 = {x0}\t|G'1(x0)| = {abs(criterio)}")

            if abs(criterio) >= 1:
                print("\n\tEl método de Newton-Raphson no converge\n") 
            else:
                print("\n\tEl método de Newton-Raphson converge\n")
                break
        except ValueError:
            print("\t Ingresa un número válido \n")
            
    arr = [float(x0)]
    i = 0  
    dif = 1

    print("i\tx_i\t\tf(x_i)\t\tf'(x_i)\t\tG(x_i)\t\tx_(i+1) - x_i\tvs tol")
    while dif > tol and i <= max_iter: 
        arr.append(G(arr[i]))
        dif = round(abs(arr[i + 1] - arr[i]), red)
        
        print(f"{i}\t{arr[i]}      \t{f(arr[i])}      \t{f_1(arr[i])}    \t{arr[i + 1]}    \t{dif}      \t{'Cumple' if dif <= tol else 'No cumple'}")
        i += 1

    print(f"\n*** Raiz = {arr[i]} ***")


def secante() :
    print("*** Secante ***")
    print("La función es: 0.5x^3 - 2x^2 + 3x - 90 = 0")

    print(f"Con tolerancia = {tol} y maximo de iteraciones = {max_iter}\n")
    tabular1(4)

    while True:
        x0 = float(input("\n*** Ingresa el valor inicial x0 *** \n$ "))
        x1 = float(input("\n*** Ingresa el valor inicial x1 *** \n$ "))

        fx0 = f(x0)
        fx1 = f(x1)

        if fx0 * fx1 >= 0 :
            print("\n\tEl intervalo no es valido\n")
        else :
            break

    i = 1
    dif = 1
    print("\n\ni\tx_(i-1)\t\tx_i\t\tf(x_(i-1))\tf(x_i)\t\tx_(i+1)\t\t|x_(i+1) - x_i|\tvs tol")
    while dif > tol and i <= max_iter:
        x2 = round(x1 - (fx1 * (x0 - x1)) / (fx0 - fx1), red)
        fx2 = f(x2)

        dif = round(abs(x2 - x1), red)
        print(f"{i}\t{x0}\t\t{x1}\t\t{fx0}\t\t{fx1}\t\t{x2}\t\t{dif}\t\t{'Cumple' if dif <= tol else 'No cumple'}")

        x0 = x1
        x1 = x2
        fx0 = fx1
        fx1 = fx2

        i += 1

    print(f"\n*** Raiz = {x2} ***")



def f(x) :
    return round(0.5*(x ** 3) - 2*(x ** 2) + 3*x - 90, red)

def f_1(x) :
    return round(1.5*(x ** 2) - 4*x + 3, red)

def f_2(x) :
    return round(3*x - 4, red)

def G(x) :
    return round(x - (f(x) / f_1(x)), red)

def G_1(x) :
    return round((f(x) * f_2(x)) / (f_1(x) ** 2), red)

def tabular1(inicio) :
    print("***** Tabulacion *****")
    print("x\tf(x)\t\tx\tf(x)\t\tx\tf(x)\t\tx\tf(x)\t\tx\tf(x)")

    for i in range(0, 10) :
        for j in range(0, 5) :
            x = inicio + (0.1 * i) + (0.1 * (10 * j))
            print(f"{round(x, red)}\t{round(f(x), red)}", end='     \t')
        print()

```

### Ecuaciones diferenciales

```python
red = 4
y_0 = 0.5
h = 0.1
tol = 0.0001

def euler() :
    print("*** Euler ***") 
    print("La función es: 2x^3 - y^2")
    print(f"Con h = {h}, y(0) = {y_0} y tolerancia = {tol}")

    y = y_0
    yi = 0
    x = 0

    yi = round(y + h*f2(x, y), red)
    diff = abs(round(abs(yi - y), red))
    print(f"\n{x}\t{y}\t{f2(x, y)}\t{yi}")

    print(f"\n*** Resultado = {yi} ***")


def euler_gauss() :
    print("*** Euler-Gauss ***")
    print("La función es: 2x^3 - y^2")
    print(f"Con h = {h}, y(0) = {y_0} y tolerancia = {tol}")

    y = y_0
    yi = 0
    x = 0
    yp = 0
    yc = 0

    yp = round(y + h*f2(x, y), red)
    yc = round(y + (h / 2)*(f2(x, y) + f2(despeje(yp), yp)), red)
    diff = abs(round(abs(yc - y), red))
    print(f"\n{x}\t{y}\t{f2(x, y)}\t\t{yp}\t\t{f2(x, yp)}\t\t\t{yc}")
    
    print(f"\n*** Resultado = {yc} ***")


def runge_kutta() :
    print("*** Runge-Kutta ***")
    print("La función es: 2x^3 - y^2")
    print(f"Con h = {h}, y(0) = {y_0} y tolerancia = {tol}")

    diff = 1
    y = y_0
    yi = 0
    x = 0

    k1 = round(f2(x, y), red)
    k2 = round(f2(x + h/2, y + (k1*h)/2), red)
    k3 = round(f2(x + h/2, y + (k2*h)/2), red)
    k4 = round(f2(x + h, y + k3*h), red)

    yi = round(y + (h*(k1 + 2*k2 + 2*k3 + k4)) / 6, red)
    diff = abs(round(abs(yi - y), red))
    print(f"\n{x}\t{y}\t{k1}\t{k2}\t{k3}\t{k4}\t{yi}")
    
    print(f"\n*** Resultado = {yi} ***")
    

def f2(x, y) :
    return round(2*(x**3) - y**2, red)

def despeje(y) :
    return round(math.cbrt((y**2) / 2), red)
```
