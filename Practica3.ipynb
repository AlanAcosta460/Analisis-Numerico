# Práctica 3 - Método de eliminación completa de Gauss-Jordan

```python
import numpy as np

redondeo = 3

A = np.array([
    [-3.0, 4.0, -2.0, 1.0, 6.0, -5.0, 2.0, -1.0, 3.0, -4.0],
    [4.0, -5.0, 2.0, 1.0, -3.0, 6.0, -2.0, -1.0, 3.0, -4.0],
    [4.0, -3.0, 1.0, 6.0, -5.0, -2.0, 3.0, -1.0, 2.0, 0.0],
    [-4.0, 2.0, 5.0, -1.0, 3.0, 0.0, -5.0, 6.0, -3.0, 1.0],
    [3.0, -4.0, 1.0, 5.0, -3.0, 2.0, -1.0, 6.0, -2.0, 4.0],
    [4.0, -3.0, 5.0, -1.0, 2.0, -5.0, 0.0, 6.0, -2.0, 3.0],
    [-3.0, 2.0, 5.0, -5.0, 1.0, 4.0, -1.0, 3.0, -6.0, 0.0],
    [-5.0, 2.0, -3.0, 1.0, 4.0, -2.0, 6.0, -4.0, 3.0, 0.0],
    [3.0, -3.0, 1.0, -2.0, 6.0, -4.0, 2.0, -1.0, 4.0, -3.0],
    [2.0, -4.0, 6.0, -3.0, 1.0, -2.0, 5.0, -1.0, 3.0, -5.0]
])
Ap = np.copy(A)

B = np.array([142.0, -135.0, -31.0, -59.0, -103.0, -64.0, -90.0, 161.0, 45.0, -68.0])
Bp = np.copy(B)

A_esc = np.zeros((10, 10))
B_esc = np.zeros(10)

I = ['x', 'y', 'z', 'w', 'v', 'u', 't', 's', 'r', 'q']

res = np.zeros(10)
com = np.zeros(10)

piv_prod = 1

def presentacion() :
    print("****************************************************")
    print("\n\t\t ANÁLISIS NUMÉRICO")
    print("\t\t      GRUPO: 4")
    print("  MÉTODO DE ELIMINACIÓN COMPLETA DE GAUSS-JORDAN\n")
    print("****************************************************")

    print("\n\n*** Integrantes del equipo ***")
    print("- Acosta Porcayo Alan Omar")
    print("- Alvarez González Eduardo")
    print("- Hernández Castillo Rubén")


def sistema(a, b) :
    for i in range(10) :
        ecuacion = ""
        for j in range(10) :
            ecuacion += f"{round(a[i][j], redondeo)}{I[j]}"
            if j < 9 :
                ecuacion += "\t+\t"
        ecuacion += f"\t=\t{round(b[i], redondeo)}"
        print(ecuacion)

def pivote(fila) :
    piv = 0
    pos = 0
    for i in range(10) :
        if abs(Ap[fila][i]) > piv :
            piv = abs(Ap[fila][i])
            pos = i
    return pos

def gauss_jordan() :
    for i in range(10) :
        pos = pivote(i)
        piv = Ap[i][pos]
        
        global piv_prod
        piv_prod *= piv
        
        for j in range(10) :
            Ap[i][j] /= piv
            if Ap[i][j] == -0.0 :
                Ap[i][j] = 0.0
        Bp[i] /= piv
        if Bp[i] == -0.0 :
            Bp[i] = 0.0

        for j in range(10) :
            if j == i :
                continue

            aux = Ap[j][pos]
            for k in range(10) :
                Ap[j][k] -= aux * Ap[i][k]
                if Ap[j][k] == -0.0 :
                    Ap[j][k] = 0.0
            Bp[j] -= aux * Bp[i]
            if Bp[j] == -0.0 :
                Bp[j] = 0.0

        print(F"\n\n%%%%% Iteración {i + 1}. Pivote: {round(piv, redondeo)} %%%%%")
        sistema(Ap, Bp)
        print()

def ordenar() :
    for i in range(10) :
        for j in range(10) :
            if Ap[i][j] == 1 :
                A_esc[j] = Ap[i]
                B_esc[j] = Bp[i]
                break    

def comprobacion() :
    for i in range(10) :
        suma = 0
        for j in range(10) :
            suma += A[i][j] * B_esc[j]
        com[i] = round(suma, redondeo)    

    
def main() :
    presentacion()

    print("\n\n%%%%% Sistema de ecuaciones %%%%%")
    sistema(A, B)
    print("\n\n")

    gauss_jordan()

    ordenar()
    print("\n\n%%%%% Sistema escalonado %%%%%")
    sistema(A_esc, B_esc)

    print("\n\n%%%%% Resultados %%%%%")
    for i in range(10) :
        print(f"{I[i]} = {round(B_esc[i], redondeo)}")

    comprobacion()
    print("\n\n%%%%% Comprobación %%%%%")
    for i in range(10) :
        print(f"{com[i]} = {B[i]}")

    print("\n\nProducto de los pivotes:", round(piv_prod, redondeo))

    det = np.linalg.det(A)
    print("Determinante de la matriz A:", round(det, redondeo))
    

    print("\nGRACIAS POR USAR EL PROGRAMA!!!")


if __name__ == "__main__":
    main()
```
