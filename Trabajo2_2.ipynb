# Método de Newton-Raphson

```python
import math

v0 = 0
tol = 0.0001
R = 2
S = 10
red = 4

def presentacion() :
    print("****************************************************")
    print("\n\t\t ANÁLISIS NUMÉRICO")
    print("\t\t      GRUPO: 4")
    print("\t\tMÉTODO DE BISECCIÓN\n")
    print("****************************************************")

    print("\n\n*** Integrantes del equipo ***")
    print("- Acosta Porcayo Alan Omar")
    print("- Alvarez González Eduardo")
    print("- Hernández Castillo Rubén")

def f(v) :
    return round(v / 2 - math.exp(v / 10) - 2, red)

def f_1(v) :
    return round(1 / 2 - math.exp(v / 10) / 10, red)

def f_2(v) :
    return round(- math.exp(v / 10) / 100, red)

def G(v) :
    return round(v - (f(v) / f_1(v)), red)

def G_1(v) :
    return round((f(v) * f_2(v)) / (f_1(v) ** 2), red)

def newton_raphson() : 
    arr = [float(v0)]
    i = 0  
    dif = 1

    print("i\tx_i\t\tf(x_i)\t\tf'(x_i)\t\tG(x_i)\t\tx_(i+1) - x_i\tvs tol")
    while dif > tol: 
        arr.append(G(arr[i]))
        dif = round(abs(arr[i + 1] - arr[i]), red)
        
        print(f"{i}\t{arr[i]}      \t{f(arr[i])}      \t{f_1(arr[i])}    \t{arr[i + 1]}  \t{dif}      \t{'Cumple' if dif <= tol else 'No cumple'}")
        i += 1

    print(f"\n*** Raiz = {arr[i]} ***")

    
def main() :
    presentacion()
    
    print()
    print("### Bienvenido al programa de Newton-Raphson ###")

    print("La función es: f(V) = V/R - e^(V/S) - 2 = 0")
    print(f"Con R = {R}, S = {S} y tolerancia = {tol}")

    global v0
    while True:
        try:
            v0 = float(input("\n*** Ingresa el valor inicial *** \n$ "))
            
            print("\n\n*** Prueba de Convergencia ***")
            print(f"|G_1(v0)| = {abs(G_1(v0))}")
            if G_1(v0) >= 1:
                print("\n\tEl método de Newton-Raphson no converge\n") 
            else:
                print("\n\tEl método de Newton-Raphson converge\n")
                break
        except ValueError:
            print("\t Ingresa un número válido \n")

    print("*** Proceso de Newton-Raphson ***")
    newton_raphson()
    
    print("\nGRACIAS POR USAR EL PROGRAMA!!!")


if __name__ == "__main__":
    main()
```
