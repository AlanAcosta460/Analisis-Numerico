# Práctica 2 - Método de Newton-Raphson

```python
import math

x0 = 0
tol = 0.0001
red = 4
max_iter = 20
incremento = 0.1
inicio = 4

def presentacion() :
    print("****************************************************")
    print("\n\t\t ANÁLISIS NUMÉRICO")
    print("\t\t      GRUPO: 4")
    print("\t\tMÉTODO DE NEWTON RAPHSON\n")
    print("****************************************************")

    print("\n\n*** Integrantes del equipo ***")
    print("- Acosta Porcayo Alan Omar")
    print("- Alvarez González Eduardo")
    print("- Hernández Castillo Rubén")

def f(x) :
    return round(0.5*(x ** 3) - 2*(x ** 2) + 3*x - 90, red)

def f_1(x) :
    return round(1.5*(x ** 2) - 4*x + 3, red)

def f_2(x) :
    return round(3*x - 4, red)

def G(x) :
    return round(x - (f(x) / f_1(x)), red)

def G_1(x) :
    return round((f(x) * f_2(x)) / (f_1(x) ** 2), red)

def tabular() :
    print("***** Tabulacion *****")
    print("x\tf(x)\t\tx\tf(x)\t\tx\tf(x)\t\tx\tf(x)\t\tx\tf(x)")

    for i in range(0, 10) :
        for j in range(0, 5) :
            x = inicio + (incremento * i) + (incremento * (10 * j))
            print(f"{round(x, red)}\t{round(f(x), red)}", end='     \t')
        print()

def newton_raphson() : 
    print("*** Proceso de Newton-Raphson ***")

    arr = [float(x0)]
    i = 0  
    dif = 1

    print("i\tx_i\t\tf(x_i)\t\tf'(x_i)\t\tG(x_i)\t\tx_(i+1) - x_i\tvs tol")
    while dif > tol and i <= max_iter: 
        arr.append(G(arr[i]))
        dif = round(abs(arr[i + 1] - arr[i]), red)
        
        print(f"{i}\t{arr[i]}      \t{f(arr[i])}      \t{f_1(arr[i])}    \t{arr[i + 1]}    \t{dif}      \t{'Cumple' if dif <= tol else 'No cumple'}")
        i += 1

    print(f"\n*** Raiz = {arr[i]} ***")

    
def main() :
    presentacion()
    
    print()
    print("### Bienvenido al programa de Newton-Raphson ###")

    print("La función es: 0.5x^3 - 2x^2 + 3x - 90 = 0")
    print(f"Con tolerancia = {tol}, maximo de iteraciones = {max_iter}\n")

    tabular()

    global x0
    while True:
        try:
            x0 = float(input("\n*** Ingresa el valor inicial *** \n$ "))
            
            criterio = G_1(x0)
            print("\n\n*** Prueba de Convergencia ***")
            print(f"x0 = {x0}\t|G_1(x0)| = {abs(criterio)}")

            if abs(criterio) >= 1:
                print("\n\tEl método de Newton-Raphson no converge\n") 
            else:
                print("\n\tEl método de Newton-Raphson converge\n")
                break
        except ValueError:
            print("\t Ingresa un número válido \n")
    
    newton_raphson()
    
    print("\nGRACIAS POR USAR EL PROGRAMA!!!")


if __name__ == "__main__":
    main()
```
